```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
from numpy import loadtxt
from xgboost import XGBClassifier
```

```python
data = pd.read_excel('data.xlsx')

print(data)
```

```python


# Extract the relevant rows and columns
labels = data.iloc[1, 2:].values
features = data.iloc[4:, 2:].astype(float).T.values

# Encode the labels (M=0, F=1)
label_encoder = LabelEncoder()
encoded_labels = label_encoder.fit_transform(labels)

# Reduce the number of features by selecting a subset
subset_features = features[:, :500]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(subset_features, encoded_labels, test_size=0.2, random_state=42)

# Create and train the XGBoost model with fewer boosting rounds
model = xgb.XGBClassifier(n_estimators=50, use_label_encoder=False, eval_metric='logloss')
model.fit(X_train, y_train)

# Make predictions and evaluate the model
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print(f"Model accuracy: {accuracy}")
```

```python
def test(nums, num):
    labels = data.iloc[1, 2:].values
    features = data.iloc[4:, 2:].astype(float).T.values
    
    # Encode the labels (M=0, F=1)
    label_encoder = LabelEncoder()
    encoded_labels = label_encoder.fit_transform(labels)
    subset_features = features[:, num:num+1]
    # Reduce the number of features by selecting a subset
    
    
    for i in range(len(nums)):
        for x in nums:
            temp = features[:, x:x+1]
            new_array = []
            for f in range(len(subset_features)):
                concat = np.concatenate((subset_features[f], temp[f]))  # Pass arrays as a tuple
                new_array.append(concat)
                

                
            subset_features = new_array[:]
            
    
    
    
    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(subset_features, encoded_labels, test_size=0.4, random_state=10)
    
    # Create and train the XGBoost model with fewer boosting rounds
    model = xgb.XGBClassifier(n_estimators=50, use_label_encoder=False, eval_metric='logloss')
    model.fit(X_train, y_train)
    
    # Make predictions and evaluate the model
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    
    

    return accuracy





    
```
